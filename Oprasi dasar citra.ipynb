```python
# Mengimpor Library Matplotlib dan diberi inisial dengan nama plt
import matplotlib.pyplot as plt
# merupakan magic command pada Jupyter Notebook atau environment lainnya yang mendefinisikan output grafik dari matplotlib langsung di dalam notebook
%matplotlib inline
# Di bawah bermaksud mengambil berbagai library dari sciencekit image
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert
# mengiport library numpy dengan inisial np
import numpy as np
```

```python
# Mengimpor modul data dari pustaka scikit-image
from skimage import data

# Memuat gambar koin ke dalam variabel coinsImage
coinsImage = data.coins()

# Memuat gambar jam ke dalam variabel clockImage
clockImage = data.clock()

# Membuat salinan gambar koin dan jam
coinsCropped = coinsImage.copy()
clockCropped = clockImage.copy()

# Memangkas gambar koin menjadi ukuran 256x256 pixel
coinsCropped = coinsCropped[0:256, 64:320]

# Memangkas gambar jam menjadi ukuran 192x192 pixel
clockCropped = clockCropped[64:256, 128:320]

# Menampilkan dimensi gambar asli dan gambar yang sudah dipangkas
print('Coins Original Shape : ', coinsImage.shape)
print('Coins Cropped Shape  : ', coinsCropped.shape)
print('Clock Original Shape : ', clockImage.shape)
print('Clock Cropped Shape  : ', clockCropped.shape)

# Membuat plot 2x2 untuk menampilkan gambar asli dan yang sudah dipangkas
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(coinsImage)
ax[0].set_title("Citra Input 1")

ax[1].imshow(clockImage, cmap='gray')
ax[1].set_title('Citra Input 2')

ax[2].imshow(coinsCropped)
ax[2].set_title("Citra Output 1")

ax[3].imshow(clockCropped, cmap='gray')
```

```python
# Mengimport fungsi invert dari modul skimage.util yang berguna untuk membalikkan warna dari gambar
inv = invert(coinsCropped)
# Menampilkan bentuk citra input dan output
print('Shape Input : ', coinsCropped.shape)
print('Shape Output : ',inv.shape)
# Membuat gambar 2x2 yang akan menampilkan citra input dan output serta histogram input dan output
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()
# Menampilkan citra input
ax[0].imshow(coinsCropped)
ax[0].set_title("Citra Input")
# Menampilkan histogram citra input
ax[1].hist(coinsCropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')
# Menampilkan citra output dari hasil inversi
ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')
# Menampilkan histogram citra output (hasil inversi)
ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```

```python
# membuat copy dari citra "clockCropped" dan mengubah tipe datanya menjadi float
copyClock = clockCropped.copy().astype(float) 
# mendapatkan dimensi matriks dari citra copyClock
m1,n1 = copyClock.shape 
# membuat array kosong dengan dimensi yang sama seperti citra "clockCropped"
output1 = np.empty([m1, n1]) 
# melakukan perulangan untuk setiap baris pada citra
for baris in range(0, m1-1): 
    for kolom in range(0, n1-1): # melakukan perulangan untuk setiap kolom pada citra
        a1 = baris
        b1 = kolom
        # menambahkan nilai 100 pada setiap piksel pada citra "clockCropped" dan disimpan ke dalam array "output1"
        output1[a1, b1] = copyClock[baris, kolom] + 100 
# membuat figure dan axes untuk menampilkan citra dan histogram        
fig, axes = plt.subplots(2, 2, figsize=(12, 12)) 
ax = axes.ravel()
# menampilkan citra "clockCropped" pada axes pertama
ax[0].imshow(clockCropped, cmap='gray') 
ax[0].set_title("Citra Input")
# menampilkan histogram dari citra "clockCropped" pada axes kedua
ax[1].hist(coinsCropped.ravel(), bins=256) 
ax[1].set_title('Histogram Input')
# menampilkan citra yang sudah ditambahkan nilai 100 pada axes ketiga
ax[2].imshow(output1, cmap='gray') 
ax[2].set_title('Citra Output (Brightness)')
# menampilkan histogram dari citra yang sudah ditambahkan nilai 100 pada axes keempat
ax[3].hist(output1.ravel(), bins=256) 
ax[3].set_title('Histogram Output')
```
